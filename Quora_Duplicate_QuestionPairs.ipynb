```python
df = df.drop(columns=['id','qid1','qid2'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv =  CountVectorizer(max_features=3000)
```

```python
questions = list(df['question1'])+list(df['question2'])
q1_arr,q2_arr = np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python
temp_1 =  pd.DataFrame(q1_arr,index = df.index )
temp_2 = pd.DataFrame(q2_arr, index = df.index)
temp_df = pd.concat([temp_1,temp_2],axis = 1)
```

```python
temp_df
```

```python
temp_df.shape
```

```python
temp_df['is_duplicate'] = df['is_duplicate']
```

```python
temp_df.head()
```

```python
temp_df.shape
```

```python
x = temp_df.iloc[:,:-1]
y = temp_df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x , y, test_size = 0.2, random_state =101)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import accuracy_score,precision_score
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred = rfc.predict(x_test)

print(accuracy_score(y_pred, y_test))
print(precision_score(y_pred,y_test))
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(x_train.to_numpy(),y_train)
y_pred = xgb.predict(x_test.to_numpy())

print(accuracy_score(y_pred, y_test))
print(precision_score(y_pred,y_test))
```
